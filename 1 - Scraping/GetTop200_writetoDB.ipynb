```python
import json
from urllib.request import urlopen

def get_top200(country):

    url = 'https://rss.itunes.apple.com/api/v1/' + country + '/podcasts/top-podcasts/all/200/explicit.json'
    with urlopen(url) as data_file:    
        data = json.load(data_file)

    results = data["feed"]    
    return results
```

```python
def writePodcastsToFile(country, results):
    fields = ['country', 'rank','name', 'id', 'releaseDate', 'Primary_Genre', 'Other_Genre' , 'Artist', "feedURL"]

    allofthem = []
    cntr = 0
    print (results["results"])
    for item in results["results"]:
        gnr = item["genres"][0]
        all_genres = ""
        all_genres_length = len(item["genres"])
        for x in range(1,all_genres_length):
            other_gnr =  item["genres"][x]
            other_name = other_gnr["name"]
            if other_name != 'Podcasts':
                all_genres +=  other_name + ","

        cntr += 1
        x = [country,cntr, item["name"], item["id"], \
            item["releaseDate"],gnr["name"], all_genres, item["artistName"], '']
        allofthem.append(x)

    return allofthem
```

```python
countries =  [  'us' ] #['us','za','uk', 'au','in','ir',] 

for c in countries:
    r = get_top200(c)
    vv = writePodcastsToFile(c, r)
    writetoDB(vv)
```
