```python
import json
from urllib.request import urlopen

def get_top200(country):

    url = 'https://rss.itunes.apple.com/api/v1/' + country + '/podcasts/top-podcasts/all/200/explicit.json'
    with urlopen(url) as data_file:    
        data = json.load(data_file)

    results = data["feed"]    
    return results
```

```python
import csv

def writePodcastsToFile(country, results):
    fields = ['country', 'rank','name', 'id', 'releaseDate', 'Primary_Genre', 'Other_Genre' , 'Artist']

    filename = "top_GB_podcasts.csv"

    allofthem = []
    cntr = 0

    for item in results["results"]:
        gnr = item["genres"][0]
        all_genres = ""
        all_genres_length = len(item["genres"])
        for x in range(1,all_genres_length):
            other_gnr =  item["genres"][x]
            other_name = other_gnr["name"]
            if other_name != 'Podcasts':
                all_genres +=  other_name + ","

        cntr += 1
        x = [country,cntr, item["name"], item["id"],  item["releaseDate"],gnr["name"], all_genres, item["artistName"]]
        allofthem.append(x)


    with open(filename, 'a', newline ='',encoding="utf-8") as csvfile: 
        writer = csv.writer(csvfile) 
        writer.writerow(fields) 
        writer.writerows(allofthem) 
```

```python
countries =  [  'gb' ] #['us','za','uk', 'au','in','ir',] 

for c in countries:
    r = get_top200(c)
    writePodcastsToFile(c, r)
```

```python
import json
from urllib.request import urlopen
import csv
import re 
#https://itunes.apple.com/us/rss/customerreviews/id=1222114325/json  - This URL gets only the first page which has 50 reviews

#Use this URL to find more pages of review.  By altering page=1 variable, to say page=2 you can get the 2nd page of 50 reviews
#
#findreviews = "https://itunes.apple.com/us/rss/customerreviews/page=1/id=" + qq + "/sortby=mostrecent/json"
#
 
def findReviews(pageNum , podcastList):
    
    filename = "podcast_reviews.csv"
    rows = []
    for i in podcastList: 
        country_code = i[0]
        podcast_id = i[1]
        rv = "https://itunes.apple.com/" + \
                country_code + "/rss/customerreviews/page=" + \
                str(pageNum) +"/id="+ str(podcast_id) +"/sortby=mostrecent/json"
        
        #print (rv)
        try:
            with urlopen(rv) as data_file:    
                reviewdata = json.load(data_file)
                reviewdatafeed = reviewdata["feed"] 

                try:
                    qq = reviewdatafeed["entry"]
                    for eachitem in qq:
                        rvwHeading=re.sub(r'[^A-Za-z0-9 \.]+', '', str(eachitem["title"]["label"])) 
                        rvwText= re.sub(r'[^A-Za-z0-9 \.]+', '',str(eachitem["content"]["label"]))
                        rvwRating = str(eachitem["im:rating"]["label"])
                        x = [podcast_id, country_code, rvwHeading, rvwText,rvwRating]
                        rows.append(x)
                except Exception as err:
                    pass
        except Exception as err:
            pass

        with open(filename, 'a', newline ='',encoding="utf-8") as csvfile: 
                writer = csv.writer(csvfile)
                for row in rows:
                    writer.writerow(row)
```
